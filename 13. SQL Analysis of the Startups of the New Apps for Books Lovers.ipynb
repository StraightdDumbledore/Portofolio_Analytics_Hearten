## Tentukan penerbit yang telah menerbitkan buku terbanyak dengan jumlah lebih dari 50 halaman

```python
a(''' SELECT publisher, count(publisher) as publ
FROM publishers
LEFT JOIN  books on publishers.publisher_id = books.publisher_id
WHERE books.num_pages > 50
GROUP BY publishers.publisher
ORDER BY publ desc''', 
 con = engine)
```

### Kesimpulan awal 3
- Publisher bernama Penguin Books memiliki 42 buku yang memiliki 50 halaman lebih 
- Disusul dengan publisher bernama Vintage yang memiliki 31 buku dengan 50 halaman lebih


## Tentukan penulis dengan rating rata-rata buku tertinggi: temukan buku dengan rating minimal 50.

```python
a('''SELECT authors.author, avg(ratings.rating) as avg_rating, count(ratings.rating) as count_rating
FROM books 
LEFT JOIN authors on books.author_id = authors.author_id
LEFT JOIN ratings on books.book_id = ratings.book_id
GROUP BY authors.author
HAVING count(ratings.rating) >= 50
ORDER BY count_rating desc, avg_rating desc
''',
  con = engine)
```

## Hitung jumlah rata-rata teks ulasan di antara pengguna yang memberi rating lebih dari 50 buku.

```python
a('''SELECT count(rating), username
FROM ratings
GROUP BY username 
HAVING count(rating) > 50''',
 con = engine)
```

#### Notes 
- Untuk menemukan jumlah rata rata teks review di antara pengguna
- Pertama Membuat tabel yang berisi pengguna yang sudah memberikan rating kepada buku yang lebih dari 50 kali 

```python
a('''SELECT username, count(text) as txt_count
FROM reviews
GROUP by username 
ORDER BY txt_count desc
''',
 con = engine)
```

#### Notes
- Membuat tabel ke dua yang berisi username dan text yang di count, sebagai paramater rata rata review yang di buat oleh pengguna.
- Contoh susan85 sudah melakukan review sebanyak 29 kali

```python
a('''WITH rate_50 as (
    SELECT count(rating) as rating_count, username
    FROM ratings
    GROUP BY username 
    HAVING count(rating) > 50
)
, main as (
    SELECT reviews.username, count(reviews.text) as review_count
    FROM reviews
    INNER JOIN rate_50 on reviews.username = rate_50.username
    GROUP BY reviews.username 
    
)
select ROUND (avg(review_count)) as avg_txt_review
from main''',
  con = engine)
```

#### Notes
- Penggabungan table yang di namakan CTE (Common Table Expression)
- Sebagai contoh melakukan table yang sudah di declare seperti yang di lakukan bernama rate_50

#### Kesimpulan awal 
- Buku yang sudah mendapatkan rating lebih dari 50 kali mempunyai jumlah rata rata review(txt_review) sebanyak 24 kali, atau 24 review oleh pengguna.
